```python
from ml_framework.data.loaders import HubSpotDataLoader
from ml_framework.data.transformers import (
    EmployeeRangeOrdinalEncoder,
    IndustryEncoder,
    MissingValueHandler,
    FeatureEngineer
)
```

```python
# 1. Load dataset with validation + profiling
loader = HubSpotDataLoader(
    customers_path="../../../Data/customers.csv",
    noncustomers_path="../../../Data/noncustomers.csv",
    usage_path="../../../Data/usage_actions.csv"
)


df = loader.load_data()
X, y = loader.get_features_and_target(df)

# 2. Feature engineering pipeline
X = EmployeeRangeOrdinalEncoder().transform(X)
ind = IndustryEncoder(min_frequency=10)
X = ind.fit_transform(X)
X = MissingValueHandler(strategy="median").fit_transform(X)
X = FeatureEngineer(scaling_method="standard").fit_transform(X)

print("Final shape:", X.shape)
print("Target distribution:", y.value_counts())
```
